```
height, width = 384, 512

def get_features(filename):
    img = cv2.resize(cv2.imread(filename), (width, height), cv2.INTER_LINEAR)/255
    return img

def get_label(filename):
    index = int(re.findall("([0-9]+)", filename)[0])
    for label in range(32):
        if index in range(datainfo.loc[label, 'Start'], datainfo.loc[label, 'End']+1):
            return label
    return float('NaN')
```

## Modeling

```
from keras.models import Model
from keras.layers import Input, Dense, Conv2D, Flatten, MaxPooling2D, Dropout

def classifier(input_shape):
    input_tensor = Input(shape=input_shape)
    conv = Conv2D(64, (3, 3), activation='relu', strides=(2, 2), padding='same')(input_tensor)
    conv = Conv2D(64, (3, 3), activation='relu', strides=(2, 2), padding='same')(conv)
    conv = MaxPooling2D()(conv)
    conv = Conv2D(128, (3, 3), activation='relu', strides=(2, 2), padding='same')(conv)
    conv = Conv2D(128, (3, 3), activation='relu', strides=(2, 2), padding='same')(conv)
    conv = MaxPooling2D()(conv)
    conv = Dropout(0.1)(conv)
    flat = Flatten()(conv)
    dense = Dense(2048, activation='relu')(flat)
    dense = Dense(256, activation='relu')(dense)
    # 32 classes:
    output_tensor = Dense(32, activation='softmax')(dense)
    model = Model(input_tensor, output_tensor)
    return model
```

## Autoencoder

```
from keras.models import Model
from keras.layers import Input, Dense, Conv2D, Flatten, Reshape, Conv2DTranspose, UpSampling2D

def autoencoder(input_shape):
    h, w, d = input_shape
    input_tensor = Input(shape=input_shape)
    # encoder
    encoder = Conv2D(16, (5, 5), activation='relu', strides=1, padding='same')(input_tensor)
    encoder = MaxPooling2D()(encoder)
    encoder = Conv2D(32, (3, 3), activation='relu', strides=1)(encoder)
    # decoder
    decoder = Conv2DTranspose(16, (3, 3), activation='relu', strides=1)(encoder)
    decoder = UpSampling2D((2, 2))(decoder)
    output_tensor = Conv2DTranspose(d, (5, 5), activation='sigmoid', strides=1, padding='same')(decoder)
    model = Model(input_tensor, output_tensor)
    return model
```

## CNN with Autoencoder

```
def AECLF(input_shape, autoencoder, main_model=None):
    input_tensor = Input(shape=input_shape)
    autoencoder.trainable = False
    autoencoder = autoencoder(input_tensor)
    if main_model == None:
        clf = classifier(input_shape)(autoencoder)
    else:
        clf = main_model(autoencoder)
    return Model(input=input_tensor, output=clf)
```
